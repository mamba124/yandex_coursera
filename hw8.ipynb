```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
data = pd.read_csv('C:/Tomato/close_prices.csv')
data.head()
```

```python
y_train = data.iloc[0]
X_train = data.iloc[:,1:]
clf = PCA(n_components = 10)
#np.seterr(divide='ignore', invalid='ignore')
clf.fit(X_train)
```

```python
print(X_train)
```

```python
clf.explained_variance_ratio_
```

```python
predict = clf.transform(X_train)
first = predict[:,0]
print(first)
```
